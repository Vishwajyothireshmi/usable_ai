Clustering based on likert questions

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
```

Finding patterns in how people respond to these mental health and usage questions.

Grouping people into clusters who respond similarly.



```python
# Select relevant columns for clustering (Likert scale questions and usage)
columns_to_use = ['how_often_do_you_find_yourself_using_social_media_without_a_specific_purpose',
 'how_often_do_you_get_distracted_by_social_media_when_you_are_busy_doing_something'
  ,'do_you_feel_restless_if_you_havent_used_social_media_in_a_while'
 ,'on_a_scale_of_1_to_5_how_easily_distracted_are_you'
 ,'on_a_scale_of_1_to_5_how_much_are_you_bothered_by_worries'
 ,'do_you_find_it_difficult_to_concentrate_on_things'
 ,'on_a_scale_of_15_how_often_do_you_compare_yourself_to_other_successful_people_through_the_use_of_social_media'
 ,'following_the_previous_question_how_do_you_feel_about_these_comparisons_generally_speaking'
 ,'how_often_do_you_look_to_seek_validation_from_features_of_social_media'
 ,'how_often_do_you_feel_depressed_or_down'
 ,'on_a_scale_of_1_to_5_how_frequently_does_your_interest_in_daily_activities_fluctuate'
 ,'on_a_scale_of_1_to_5_how_often_do_you_face_issues_regarding_sleep']
data = df[columns_to_use].dropna()
```

```python
# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

```python
# Find optimal number of clusters using the Elbow Method
inertia = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

# Plot Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.show()
```

```python
# optimal k=3 based on elbow plot
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init='auto')
clusters = kmeans.fit_predict(scaled_data)

# Add cluster labels to the dataframe
data['Cluster'] = clusters
```

```python
# Reduce to 2D for visualization using PCA
pca = PCA(n_components=2)
pca_components = pca.fit_transform(scaled_data)

# Visualize clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(x=pca_components[:, 0], y=pca_components[:, 1], hue=clusters, palette='Set2')
plt.title('K-Means Clustering (PCA 2D Projection)')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
# Add clusters to original dataset
df_clustered = df.copy()
df_clustered = df_clustered[columns_to_use]  # Q9 to Q20
df_clustered['Cluster'] = clusters

# Calculate the mean response for each question per cluster
cluster_means = df_clustered.groupby('Cluster').mean()
print(cluster_means)
```

## Model Training

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

## Random Forest

```python
mental_health_cols = ['how_often_do_you_find_yourself_using_social_media_without_a_specific_purpose',
 'how_often_do_you_get_distracted_by_social_media_when_you_are_busy_doing_something'
  ,'do_you_feel_restless_if_you_havent_used_social_media_in_a_while'
 ,'on_a_scale_of_1_to_5_how_easily_distracted_are_you'
 ,'on_a_scale_of_1_to_5_how_much_are_you_bothered_by_worries'
 ,'do_you_find_it_difficult_to_concentrate_on_things'
 ,'on_a_scale_of_15_how_often_do_you_compare_yourself_to_other_successful_people_through_the_use_of_social_media'
 ,'how_often_do_you_feel_depressed_or_down'
 ,'on_a_scale_of_1_to_5_how_frequently_does_your_interest_in_daily_activities_fluctuate'
 ,'on_a_scale_of_1_to_5_how_often_do_you_face_issues_regarding_sleep']
features = df[mental_health_cols]

df['mental_distress_score'] = df[mental_health_cols].mean(axis=1)
df['high_distress'] = df['mental_distress_score'].apply(lambda x: 1 if x >= 3 else 0)
target = df['high_distress']
```

```python
# Drop rows with missing values, if any
features = features.dropna()
target = target.loc[features.index]
```

```python
# Step 3: Train-test split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
# Step 4: Train Random Forest

rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

print("Random Forest Classification Report")
print(classification_report(y_test, y_pred_rf))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Get confusion matrix
cm = confusion_matrix(y_test, y_pred_rf)

# Plot heatmap
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Not Depressed (0)', 'Depressed (1)'],
            yticklabels=['Not Depressed (0)', 'Depressed (1)'])

plt.title('Confusion Matrix - Random Forest')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
importances = pd.Series(rf.feature_importances_, index=mental_health_cols)
plt.figure(figsize=(8,6))
importances.sort_values().plot(kind='barh', color='skyblue')
plt.title('Feature Importance - Random Forest')
plt.show()
```

## Platform-Specific Effects

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Detect correct platform column
platform_col = [col for col in df.columns if "platform" in col.lower() and "use" in col.lower()]
platform_col = platform_col[0] if platform_col else None
print("Detected platform column:", platform_col)

# Step 2: Generate binary flags for each platform
platforms = ['Facebook', 'Instagram', 'Twitter', 'YouTube', 'Snapchat', 'Pinterest', 'TikTok', 'Reddit', 'Discord']
for platform in platforms:
    df[f'uses_{platform.lower()}'] = df[platform_col].fillna('').apply(lambda x: int(platform.lower() in x.lower()))

# Step 3: Cleaned column references for depression and validation seeking
score_cols = {
    'Depression': 'how_often_do_you_feel_depressed_or_down',
    'Validation Seeking': 'how_often_do_you_look_to_seek_validation_from_features_of_social_media'
}

# Step 4: Calculate averages
platform_effects = {}
for platform in platforms:
    users = df[df[f'uses_{platform.lower()}'] == 1]
    non_users = df[df[f'uses_{platform.lower()}'] == 0]
    stats = {
        'Depression (users)': users[score_cols['Depression']].mean(),
        'Depression (non-users)': non_users[score_cols['Depression']].mean(),
        'Validation (users)': users[score_cols['Validation Seeking']].mean(),
        'Validation (non-users)': non_users[score_cols['Validation Seeking']].mean()
    }
    platform_effects[platform] = stats

# Step 5: Convert to DataFrame
platform_df = pd.DataFrame(platform_effects).T
platform_df_sorted = platform_df.sort_index()
platforms = platform_df_sorted.index.tolist()
x = np.arange(len(platforms))  # X axis positions
width = 0.2  # width of each bar

# Step 6: Plotting
fig, ax = plt.subplots(figsize=(14, 6))
plt.style.use('ggplot')

# Plot each bar group
bars1 = ax.bar(x - width, platform_df_sorted['Depression (users)'], width, label='Depression (users)', color='#4c72b0')
bars2 = ax.bar(x, platform_df_sorted['Depression (non-users)'], width, label='Depression (non-users)', color='#dd8452')
bars3 = ax.bar(x + width, platform_df_sorted['Validation (users)'], width, label='Validation (users)', color='#55a868')
bars4 = ax.bar(x + 2 * width, platform_df_sorted['Validation (non-users)'], width, label='Validation (non-users)', color='#c44e52')

# Add value labels
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}', xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3), textcoords="offset points", ha='center', fontsize=8)

[add_labels(b) for b in [bars1, bars2, bars3, bars4]]

# Final styling
ax.set_ylabel('Average Score')
ax.set_title('Depression and Validation Seeking by Platform Usage')
ax.set_xticks(x + width / 2)
ax.set_xticklabels(platforms, rotation=45)
ax.legend()
ax.set_ylim(0, 4.5)
plt.tight_layout()
plt.show()
```

## Mediation Analysis

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Step 1: Detect time-spent column
time_col = [col for col in df.columns if "time" in col.lower() and "social" in col.lower()]
if time_col:
    time_col = time_col[0]
    print("Detected time column:", time_col)
else:
    raise ValueError("Could not auto-detect time-spent column.")

# Step 2: Find correct comparison column
comparison_col = [col for col in df.columns if "compare" in col.lower()]
if comparison_col:
    comparison_col = comparison_col[0]
    print("Detected comparison column:", comparison_col)
else:
    raise ValueError("Could not auto-detect social comparison column.")

# Step 3: Use known depression column
depression_col = 'how_often_do_you_feel_depressed_or_down'

# Step 4: Prepare mediation DataFrame
df_med = df[[time_col, comparison_col, depression_col]].dropna()
df_med = df_med.rename(columns={
    time_col: 'time_spent',
    comparison_col: 'comparison',
    depression_col: 'depression'
})

# Convert time_spent (categorical) to ordinal numeric
df_med['time_spent'] = df_med['time_spent'].astype('category').cat.codes

# Step 5: Fit models for mediation
# Path A: time_spent ‚ûù comparison
model_a = ols('comparison ~ time_spent', data=df_med).fit()

# Path B & C': comparison + time_spent ‚ûù depression
model_b = ols('depression ~ comparison + time_spent', data=df_med).fit()

# Path C: total effect (time_spent ‚ûù depression)
model_c = ols('depression ~ time_spent', data=df_med).fit()

# Step 6: Compute effects
a = model_a.params['time_spent']
b = model_b.params['comparison']
c_prime = model_b.params['time_spent']
c_total = model_c.params['time_spent']
indirect_effect = a * b

# Step 7: Report
print("Mediation Analysis Summary")
print("--------------------------")
print(f"Path A (time ‚ûù comparison):     {a:.4f}")
print(f"Path B (comparison ‚ûù depression): {b:.4f}")
print(f"Indirect effect (A √ó B):        {indirect_effect:.4f}")
print(f"Direct effect (C'):             {c_prime:.4f}")
print(f"Total effect (C):               {c_total:.4f}")
```
